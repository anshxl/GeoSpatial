```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Get base directory
BASE = Path.cwd().parent

# Load the data
data_path = BASE / 'outputs' / 'streaming_predictions.csv'
data = pd.read_csv(data_path)
data.head()
```

```python
# Get accuracy
accuracy = sum(data['true_label'] == data['pred_label']) / len(data)
# Print accuracy
print(f"Accuracy: {accuracy:.2%}")
```

```python
# Group by true label and print average confidence
avg_confidence = data.groupby('true_label')['confidence'].mean()

# # Print average confidence
# print("Average confidence by true label:")
# print(avg_confidence)

# Plot the average confidence
plt.figure(figsize=(10, 6))
avg_confidence.plot(kind='bar', color='skyblue')
plt.title('Average Confidence by True Label')
plt.xlabel('True Label')
plt.ylabel('Average Confidence')
plt.xticks(rotation=45)
plt.ylim((avg_confidence.min() - 0.05, avg_confidence.max() + 0.05))
plt.tight_layout()
```
